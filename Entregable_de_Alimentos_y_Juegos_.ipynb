```python
print("="*60)
print("ABSTRACCIÓN - ALIMENTOS")
print("="*60)

class Alimento:
    def __init__(self, nombre, tipo, calorias, precio):
        self.nombre = nombre
        self.tipo = tipo
        self.calorias = calorias
        self.precio = precio

    def consumir(self):
        print(f"Consumiendo {self.nombre}")

    def mostrar_informacion(self):
        print(f"Alimento: {self.nombre}")
        print(f"Tipo: {self.tipo}")
        print(f"Calorías: {self.calorias} kcal")
        print(f"Precio: ${self.precio}")

pizza = Alimento("Pizza Margarita", "Comida rápida", 266, 45.000)
ensalada = Alimento("Ensalada César", "Vegetales", 150, 13.000)

pizza.mostrar_informacion()
pizza.consumir()
print()
ensalada.mostrar_informacion()
ensalada.consumir()

print("\n\n")
```

```python
print("="*60)
print("ENCAPSULAMIENTO - ALIMENTOS")
print("="*60)

class Alimento:
    def __init__(self, nombre, tipo, calorias, precio):
        self.__nombre = nombre
        self.__tipo = tipo
        self.__calorias = calorias
        self.__precio = precio
        self.__disponible = True

    def get_nombre(self):
        return self.__nombre

    def get_precio(self):
        return self.__precio

    def get_disponible(self):
        return self.__disponible

    def set_precio(self, nuevo_precio):
        if nuevo_precio > 0:
            self.__precio = nuevo_precio
            print(f"Precio actualizado a ${nuevo_precio}")
        else:
            print("El precio debe ser mayor a 0")

    def set_disponible(self, estado):
        self.__disponible = estado
        if estado:
            print(f"{self.__nombre} ahora está disponible")
        else:
            print(f"{self.__nombre} no está disponible")

    def mostrar_informacion(self):
        estado = "Disponible" if self.__disponible else "No disponible"
        print(f"{self.__nombre} - ${self.__precio} - {estado}")

hamburguesa = Alimento("Hamburguesa", "Comida rápida", 540, 7.50)
hamburguesa.mostrar_informacion()
hamburguesa.set_precio(8.00)
hamburguesa.set_disponible(False)
hamburguesa.mostrar_informacion()

print("\n\n")
```

```python
print("="*60)
print("CÓDIGO 3: HERENCIA - ALIMENTOS")
print("="*60)

class Alimento:
    def __init__(self, nombre, tipo, calorias):
        self.nombre = nombre
        self.tipo = tipo
        self.calorias = calorias

    def consumir(self):
        print(f"Consumiendo {self.nombre}")

    def mostrar_informacion(self):
        print(f"Alimento: {self.nombre}, Tipo: {self.tipo}, Calorías: {self.calorias}")

class Fruta(Alimento):
    def __init__(self, nombre, calorias, vitaminas, temporada):
        super().__init__(nombre, "Fruta", calorias)
        self.vitaminas = vitaminas
        self.temporada = temporada

    def pelar(self):
        print(f"Pelando {self.nombre}")

    def mostrar_informacion(self):
        super().mostrar_informacion()
        print(f"Vitaminas: {self.vitaminas}, Temporada: {self.temporada}")

class Verdura(Alimento):
    def __init__(self, nombre, calorias, minerales, color):
        super().__init__(nombre, "Verdura", calorias)
        self.minerales = minerales
        self.color = color

    def lavar(self):
        print(f"Lavando {self.nombre}")

    def cocinar(self):
        print(f"Cocinando {self.nombre} al vapor")

class Proteina(Alimento):
    def __init__(self, nombre, calorias, gramos_proteina, origen):
        super().__init__(nombre, "Proteína", calorias)
        self.gramos_proteina = gramos_proteina
        self.origen = origen

    def cocinar_parrilla(self):
        print(f"Cocinando {self.nombre} a la parrilla")

mango = Fruta("Mango", 60, "A, C", "Verano")
mango.mostrar_informacion()
mango.pelar()
mango.consumir()

print()

espinaca = Verdura("Espinaca", 23, "Hierro, Calcio", "Verde")
espinaca.lavar()
espinaca.cocinar()
espinaca.consumir()

print()

pollo = Proteina("Pechuga de pollo", 165, 31, "Animal")
pollo.cocinar_parrilla()
pollo.consumir()

print("\n\n")
```

```python
print("="*60)
print("CÓDIGO 4: POLIMORFISMO - ALIMENTOS")
print("="*60)

class Alimento:
    def __init__(self, nombre, tipo, calorias):
        self.nombre = nombre
        self.tipo = tipo
        self.calorias = calorias

    def preparar(self):
        print(f"Preparando {self.nombre}")

    def consumir(self):
        print(f"Consumiendo {self.nombre}")

class Fruta(Alimento):
    def __init__(self, nombre, calorias):
        super().__init__(nombre, "Fruta", calorias)

    def preparar(self):
        print(f"Lavando y cortando {self.nombre} en trozos")

class Carne(Alimento):
    def __init__(self, nombre, calorias):
        super().__init__(nombre, "Carne", calorias)

    def preparar(self):
        print(f"Marinando y cocinando {self.nombre} a fuego medio")

class Bebida(Alimento):
    def __init__(self, nombre, calorias):
        super().__init__(nombre, "Bebida", calorias)

    def preparar(self):
        print(f"Sirviendo {self.nombre} en un vaso con hielo")

class Postre(Alimento):
    def __init__(self, nombre, calorias):
        super().__init__(nombre, "Postre", calorias)

    def preparar(self):
        print(f"Decorando {self.nombre} con crema y frutas")


lista_alimentos = [
    Fruta("Fresa", 32),
    Carne("Bistec", 271),
    Bebida("Jugo de naranja", 112),
    Postre("Tiramisu", 450)
]


print("Preparando diferentes alimentos:")
for alimento in lista_alimentos:
    alimento.preparar()

print("\n\n")
```

```python
print("="*60)
print("ABSTRACCIÓN - JUEGOS")
print("="*60)

class Juego:
    def __init__(self, nombre, categoria, num_jugadores, edad_minima):
        self.nombre = nombre
        self.categoria = categoria
        self.num_jugadores = num_jugadores
        self.edad_minima = edad_minima

    def iniciar(self):
        print(f"Iniciando el juego {self.nombre}")

    def finalizar(self):
        print(f"Finalizando {self.nombre}")

    def mostrar_informacion(self):
        print(f"Juego: {self.nombre}")
        print(f"Categoría: {self.categoria}")
        print(f"Jugadores: {self.num_jugadores}")
        print(f"Edad mínima: {self.edad_minima} años")

ajedrez = Juego("Ajedrez", "Estrategia", 2, 6)
futbol = Juego("Fútbol", "Deporte", 22, 5)

ajedrez.mostrar_informacion()
ajedrez.iniciar()
ajedrez.finalizar()
print()
futbol.mostrar_informacion()
futbol.iniciar()

print("\n\n")
```

```python
print("="*60)
print("ENCAPSULAMIENTO - JUEGOS")
print("="*60)

class Juego:
    def __init__(self, nombre, categoria, num_jugadores):
        self.__nombre = nombre
        self.__categoria = categoria
        self.__num_jugadores = num_jugadores
        self.__puntuacion = 0
        self.__nivel = 1

    def get_nombre(self):
        return self.__nombre

    def get_puntuacion(self):
        return self.__puntuacion

    def get_nivel(self):
        return self.__nivel

    def set_puntuacion(self, puntos):
        if puntos >= 0:
            self.__puntuacion = puntos
            print(f"Puntuación actualizada: {puntos} puntos")
        else:
            print("La puntuación no puede ser negativa")

    def incrementar_puntuacion(self, puntos):
        if puntos > 0:
            self.__puntuacion += puntos
            print(f"+{puntos} puntos! Total: {self.__puntuacion}")
        else:
            print("Los puntos deben ser positivos")

    def subir_nivel(self):
        self.__nivel += 1
        print(f"¡Nivel {self.__nivel} desbloqueado!")

    def mostrar_estadisticas(self):
        print(f"Juego: {self.__nombre}")
        print(f"Nivel: {self.__nivel}")
        print(f"Puntuación: {self.__puntuacion}")

mario_kart = Juego("Mario Kart", "Carreras", 4)
mario_kart.mostrar_estadisticas()
mario_kart.incrementar_puntuacion(100)
mario_kart.incrementar_puntuacion(50)
mario_kart.subir_nivel()
mario_kart.mostrar_estadisticas()

print("\n\n")
```

```python
print("="*60)
print("HERENCIA - JUEGOS")
print("="*60)

class Juego:
    def __init__(self, nombre, categoria, num_jugadores):
        self.nombre = nombre
        self.categoria = categoria
        self.num_jugadores = num_jugadores

    def iniciar(self):
        print(f"Iniciando {self.nombre}")

    def mostrar_informacion(self):
        print(f"Juego: {self.nombre}, Categoría: {self.categoria}, Jugadores: {self.num_jugadores}")

class JuegoMesa(Juego):
    def __init__(self, nombre, num_jugadores, duracion, tiene_dado):
        super().__init__(nombre, "Juego de Mesa", num_jugadores)
        self.duracion = duracion
        self.tiene_dado = tiene_dado

    def preparar_tablero(self):
        print(f"Preparando el tablero de {self.nombre}")

    def lanzar_dado(self):
        if self.tiene_dado:
            print(f"Lanzando dado en {self.nombre}")
        else:
            print(f"{self.nombre} no usa dados")

class Videojuego(Juego):
    def __init__(self, nombre, num_jugadores, plataforma, tamaño_gb):
        super().__init__(nombre, "Videojuego", num_jugadores)
        self.plataforma = plataforma
        self.tamaño_gb = tamaño_gb

    def descargar(self):
        print(f"Descargando {self.nombre} ({self.tamaño_gb} GB) para {self.plataforma}")

    def guardar_partida(self):
        print(f"Guardando progreso de {self.nombre}")

class JuegoDeporte(Juego):
    def __init__(self, nombre, num_jugadores, duracion_partido, equipamiento):
        super().__init__(nombre, "Deporte", num_jugadores)
        self.duracion_partido = duracion_partido
        self.equipamiento = equipamiento

    def calentar(self):
        print(f"Calentando para jugar {self.nombre}")

    def mostrar_equipamiento(self):
        print(f"Equipamiento necesario: {self.equipamiento}")

monopoly = JuegoMesa("Monopoly", 4, "2-3 horas", True)
monopoly.mostrar_informacion()
monopoly.preparar_tablero()
monopoly.lanzar_dado()

print()

zelda = Videojuego("Zelda: Breath of the Wild", 1, "Nintendo Switch", 14.4)
zelda.descargar()
zelda.iniciar()
zelda.guardar_partida()

print()

basquet = JuegoDeporte("Básquetbol", 10, "40 minutos", "Balón, cancha, aros")
basquet.calentar()
basquet.mostrar_equipamiento()
basquet.iniciar()

print("\n\n")
```

```python
print("="*60)
print("CÓDIGO 8: POLIMORFISMO - JUEGOS")
print("="*60)

class Juego:
    def __init__(self, nombre, categoria):
        self.nombre = nombre
        self.categoria = categoria

    def jugar(self):
        print(f"Jugando {self.nombre}")

    def ganar(self):
        print(f"¡Has ganado en {self.nombre}!")


class JuegoCartas(Juego):
    def __init__(self, nombre):
        super().__init__(nombre, "Cartas")

    def jugar(self):
        print(f"Barajando y repartiendo cartas para {self.nombre}")

class JuegoArcade(Juego):
    def __init__(self, nombre):
        super().__init__(nombre, "Arcade")

    def jugar(self):
        print(f"Insertando moneda y presionando START en {self.nombre}")

class JuegoRol(Juego):
    def __init__(self, nombre):
        super().__init__(nombre, "Rol")

    def jugar(self):
        print(f"Creando personaje y comenzando aventura en {self.nombre}")

class JuegoEstrategia(Juego):
    def __init__(self, nombre):
        super().__init__(nombre, "Estrategia")

    def jugar(self):
        print(f"Planificando estrategia y moviendo tropas en {self.nombre}")

lista_juegos = [
    JuegoCartas("Póker"),
    JuegoArcade("Pac-Man"),
    JuegoRol("Dungeons & Dragons"),
    JuegoEstrategia("Age of Empires")
]

print("Jugando diferentes juegos:")
for juego in lista_juegos:
    juego.jugar()

print("\n" + "="*60)
```
