```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("="*70)
print("PARCIAL - ANÁLISIS DE DATOS CON PYTHON")
print("="*70)

# PREGUNTA 1: NumPy - Estadísticas básicas
print("\n" + "="*70)
print("PREGUNTA 1: Análisis estadístico con NumPy")
print("="*70)

ventas_array = np.array([120, 340, 290, 410, 150, 380])

promedio = np.mean(ventas_array)
maximo = np.max(ventas_array)
minimo = np.min(ventas_array)

print(f"Arreglo de ventas: {ventas_array}")
print(f"\nPromedio de ventas: {promedio:.2f}")
print(f"Venta máxima: {maximo}")
print(f"Venta mínima: {minimo}")

# PREGUNTA 2: NumPy - Operaciones condicionales sin ciclos
print("\n" + "="*70)
print("PREGUNTA 2: Clasificación condicional sin ciclos")
print("="*70)

clasificacion = (ventas_array > 300).astype(int)

print(f"Ventas: {ventas_array}")
print(f"Clasificación (1 si >300, 0 caso contrario): {clasificacion}")

# PREGUNTA 3: Pandas - Columnas calculadas
print("\n" + "="*70)
print("PREGUNTA 3: Creación de columnas calculadas")
print("="*70)

data = {
    'Producto': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Ventas': [120, 340, 290, 410, 150, 380],
    'Costo': [80, 200, 180, 260, 90, 240],
    'Categoria': ['Tecnología', 'Hogar', 'Tecnología', 'Hogar', 'Oficina', 'Tecnología'],
    'Mes': ['Enero', 'Enero', 'Febrero', 'Febrero', 'Marzo', 'Marzo']
}
df = pd.DataFrame(data)

df['Ganancia'] = df['Ventas'] - df['Costo']

print("Primeras filas del DataFrame:")
print(df.head())

productos_rentables = df[df['Ganancia'] > 100]
print(f"\nProductos con ganancia > 100:")
print(productos_rentables[['Producto', 'Ganancia']])


# PREGUNTA 4: Pandas - Filtros múltiples
print("\n" + "="*70)
print("PREGUNTA 4: Filtrado con múltiples condiciones")
print("="*70)

tecnologia_alto = df[(df['Categoria'] == 'Tecnología') & (df['Ventas'] > 300)]
tecnologia_alto = tecnologia_alto[['Producto', 'Ventas', 'Ganancia']]

print("Productos de Tecnología con ventas > 300:")
print(tecnologia_alto)

# PREGUNTA 5: Pandas - Agrupaciones
print("\n" + "="*70)
print("PREGUNTA 5: Análisis con groupby")
print("="*70)

resumen_mensual = df.groupby('Mes').agg({
    'Ventas': 'sum',
    'Ganancia': 'mean'
}).round(2)

print("Resumen por Mes:")
print(resumen_mensual)

# PREGUNTA 6: Matplotlib - Gráfico de barras
print("\n" + "="*70)
print("PREGUNTA 6: Visualización con gráfico de barras")
print("="*70)

plt.figure()
plt.bar(df['Producto'], df['Ventas'], color='steelblue')
plt.title('Ventas por Producto')
plt.xlabel('Producto')
plt.ylabel('Ventas ($)')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.show()

print("Gráfico de barras generado")

# PREGUNTA 7: Matplotlib - Gráfico de líneas comparativo
print("\n" + "="*70)
print("PREGUNTA 7: Comparación con gráfico de líneas")
print("="*70)

plt.figure()
plt.plot(df['Producto'], df['Ventas'], marker='o', label='Ventas', color='green')
plt.plot(df['Producto'], df['Costo'], marker='o', label='Costo', color='red')
plt.title('Ventas vs Costo por Producto')
plt.xlabel('Producto')
plt.ylabel('Monto ($)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("ANÁLISIS:")
print("- Las ventas siempre están por encima de los costos (todos rentables)")
print("- El producto D tiene la mayor separación (mayor ganancia)")
print("- Existe correlación positiva entre costos y ventas")

# PREGUNTA 8: Seaborn - Boxplot
print("\n" + "="*70)
print("PREGUNTA 8: Análisis de variabilidad con boxplot")
print("="*70)

plt.figure()
sns.boxplot(x='Categoria', y='Ventas', data=df)
plt.title('Distribución de Ventas por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Ventas ($)')
plt.grid(axis='y', alpha=0.3)
plt.show()

variabilidad = df.groupby('Categoria')['Ventas'].agg(['std', 'min', 'max'])
variabilidad['rango'] = variabilidad['max'] - variabilidad['min']
print("\nEstadísticas de variabilidad:")
print(variabilidad)

print("\nANÁLISIS:")
print("- MAYOR VARIABILIDAD: Tecnología")
print("- Tiene la desviación estándar más alta")
print("- Rango más amplio (120 a 380)")
print("- Conclusión: Tecnología tiene ventas más dispersas")

# PREGUNTA 9: Seaborn - Scatterplot
print("\n" + "="*70)
print("PREGUNTA 9: Relación Ventas-Ganancia con scatterplot")
print("="*70)

plt.figure()
sns.scatterplot(x='Ventas', y='Ganancia', hue='Categoria', data=df, s=100)
plt.title('Relación entre Ventas y Ganancia por Categoría')
plt.xlabel('Ventas ($)')
plt.ylabel('Ganancia ($)')
plt.legend(title='Categoría')
plt.grid(True, alpha=0.3)
plt.show()

correlacion = df['Ventas'].corr(df['Ganancia'])
print(f"\nCoeficiente de correlación: {correlacion:.3f}")

print("\nANÁLISIS:")
print(f"- Relación positiva fuerte (r = {correlacion:.3f})")
print("- A mayor venta, mayor ganancia")
print("- Se observa comportamiento lineal")

# PREGUNTA 10: Análisis Final y Recomendaciones
print("\n" + "="*70)
print("PREGUNTA 10: Análisis Final y Recomendaciones")
print("="*70)

producto_rentable = df.loc[df['Ganancia'].idxmax()]
print("\na) PRODUCTO MÁS RENTABLE:")
print(f"   Producto: {producto_rentable['Producto']}")
print(f"   Ganancia: ${producto_rentable['Ganancia']}")

ganancia_categoria = df.groupby('Categoria')['Ganancia'].sum().sort_values(ascending=False)
print("\nb) GANANCIA TOTAL POR CATEGORÍA:")
print(ganancia_categoria)
print(f"\n   Categoría líder: {ganancia_categoria.idxmax()}")
print(f"   Ganancia total: ${ganancia_categoria.max()}")

print("\nc) RECOMENDACIÓN COMERCIAL:")
print("\n   DECISIÓN: Invertir o mantener en Tecnología y Hogar")
print("\n   JUSTIFICACIÓN:")
print("   1. Ambas categorías generan $290 de ganancia total")
print("   2. Producto D (Hogar) es el más rentable individualmente")
print("   3. Tecnología tiene potencial de mejora (producto A bajo rendimiento)")
print("\n   ACCIÓN:")
print("   - Mantener productos de alto rendimiento (D, F)")
print("   - Investigar por qué producto A tiene bajo rendimiento")
print("   - Replicar estrategias exitosas en productos de baja venta")

```
